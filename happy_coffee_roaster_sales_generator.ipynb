    10. price_3 : The price of the order 3

    12. price_5 : The price of the order 5
    13. staff : Staff name  

```python
import random
import numpy as np
import pandas as pd
import calendar
import datetime as dt


#Random a consecutive month-range from num_of_month and year input
def random_months_range_generator(num_of_month, year):  
    months = list(range(1,13))
    start_month = random.choice(months)
    random_months = []
    counter = 0
    for counter in range(num_of_month):
        if start_month <= 12:
            random_months.append(start_month)
            start_month += 1
        elif start_month > 12:
            start_month = 1
            random_months.append(start_month)
            start_month += 1
            
    date_ranges = []
    if random_months[0] < 11: #In case of it needs to overlap the next year
        date_ranges.append("1/" + str(random_months[0]) + "/" + str(year))
        date_ranges.append(str(calendar.monthrange(2021, random_months[-1])[1]) + "/" + str(random_months[-1]) + "/" + str(year))
    else:
        date_ranges.append("1/" + str(random_months[0]) + "/" + str(year))
        date_ranges.append(str(calendar.monthrange(2021, random_months[-1])[1]) + "/" + str(random_months[-1]) + "/" + str(year + 1))
    return date_ranges

example_month_ranges = random_months_range_generator(3, 2021)
print(example_month_ranges)

```

We need to make Python recognise the date format in the example_month_ranges in "__dd/mm/yy__" not in the American format like "__mm/dd/yy__" before generating a list of dates. 

```python
#Create a list of string dates from example_month_ranges
start_time = dt.datetime.strptime(example_month_ranges[0], "%d/%m/%Y")
end_time = dt.datetime.strptime(example_month_ranges[1], "%d/%m/%Y")
date_objs = pd.date_range(start = start_time, end = end_time).to_pydatetime().tolist()
dates = []
for date in date_objs:
    dates.append(date.strftime("%d-%m-%Y"))
print(dates)
print("\n")
num_of_days = len(dates)
print("Total days =", num_of_days)
```

```python
#Create total random numbers of order each day that ranges from 5-30 orders
#The length is based on Total days
orders = random.choices(range(5, 31), k = num_of_days)
print(orders)
```

```python
#Create a list of repetitive new dates based on an order number each day
repeated_dates = sum([[d] * o for d, o in zip(dates, orders)] , [])
print(repeated_dates)
```

```python
#Check if the first order number matches the dates in repetitive_dates
print("The order number of the first day is", orders[0])
count = 0
for date in repeated_dates:
    if date == repeated_dates[0]:
        count += 1
if count == orders[0]:
    print("There are", count, "repeated dates for the first day")
else:
    print("There is something wrong in the code!")
```

After creating a list of repeated random dates, a list of random times is also needed for each order. The list of times is continuous from the time the coffe shop opens until it is closed by 16:30 PM, also the repeated times are based on the number of order each day. 

```python
#Create a function to generate a list of random times
def random_times(shour, ehour):
    string_times = []
    for num_order in orders:
        random_time_in_aday = []
        for num in range(num_order):
            random_hour = str(random.randint(shour, ehour))
            random_min = str(random.randint(0, 59))
            if len(random_min) < 2:
                random_min = "0" + random_min
            random_time = random_hour + "." + random_min
            if float(random_time) < 7.30: #Updated 16/11/2021
                random_time_in_aday.append(random.uniform(7.30, 7.59))
            elif float(random_time) > 16.30:
                random_time_in_aday.append(random.uniform(16.00, 16.30))
            else:
                random_time_in_aday.append(random_time)   
        random_time_in_aday = ["{:.2f}".format(float(item)) for item in random_time_in_aday]
        random_time_in_aday.sort()
        random_time_in_aday = [str(item) for item in random_time_in_aday] #Change a type of each item back to string in order to change "." to ":"
        random_time_in_aday = [item.replace(".", ":") for item in random_time_in_aday]
        string_times += random_time_in_aday
        random_time_in_aday.clear()
    return string_times

times = random_times(7, 16) #The 7 means the coffe shop opens at 7:30AM, and 16 means it is closed by 16:30PM. 

#Print the first 50 time samples
print("Below is the list of the first 50 sample time stamps:")
print(times[0:50]) 
print("\n")

#Check if the total repeated times is equal to the total dates
if len(times) == len(repeated_dates):
    print("The total times and dates based on the numer of order each day are equal!")
    print("There are", len(times), "time stamps and", len(repeated_dates), "dates")
else:
    print("There is something wrong in the code!")
```

Assign a random number of menus ranged from 1-5 to each order. Assume that most of the customers are alone or couple, therefore they are likely to order 1-3 menus each order. The "__random.choices__" with weight argument will help manage this.  

```python
#Create a dictionary of menus and prices, a list of menus, and a list of order number.
menu_prices= {
"Espresso shot": 30,
"Americano": 50,
"Latte": 70,
"Cappucino": 70,
"Flat white": 70,
"Mocha": 75,
"Macchiato": 75,
"Pour-over normal": 100,
"Pour-over special": 200,
"Coffee beans 200": 400,
"Coffee beans special 200": 600
}
menu = ["Espresso shot", "Americano", "Latte", "Cappucino", "Flat white", "Mocha", "Macchiato", "Pour-over normal", "Pour-over special", "Coffee beans 200", "Coffee beans special 200"]
each_order_number = [1, 2, 3, 4, 5]
#Detemine the possibility of each order number in case of most customers tend to order 1-3 menus each order
order_weights = [0.5, 0.3, 0.1, 0.05, 0.05]
#Detemine the possibility of each menu to be ordered so the pour-over coffee tend to be ordered more than the others
menu_weights = [0.05, 0.1, 0.1, 0.1, 0.1, 0.05, 0.05, 0.2, 0.15, 0.05, 0.05]

#Create a function to generate a list of random order numbers and menus in each order
each_order = []

def assign_menu(ordered_times, weights_1, weights_2):
    each_order = [] #The list of each random order number
    menus = [] #The final list of lists that contained menus of each order
    each_order_menus = [] #The list of each random order menus
    for time in times:
        each_order += random.choices(each_order_number, weights = weights_1)
    for each in each_order:
        for i in range(each):
            each_order_menus += random.choices(menu, weights = weights_2)
        menus.append(each_order_menus)
        each_order_menus = []
    return menus

menus = assign_menu(times, order_weights, menu_weights)
print("Below is the all order samples in the first day (28 orders):")
print(menus[0:28])
print("\n")
#Check if the total list of lists is equal to the times list
if len(menus) == len(times):
    print("The total list of lists is", len(menus), "which is equal to", len(times))
else:
    print("There is something wrong in the code!")
```

Next up, create a function that returns a list of total prices for each order by mapping the menu names to the prices in the dictionary. The function accepts 2 arguments

1. __all_orders__: This is the list of lists created above which is assigned to the variable __menus__
2. __menus_dict__: A dictionary contains the price for each menu, in this case the variable named __menu_prices__

```python
def assign_price(all_orders, menus_dict): 
    all_prices = []
    order_prices = []
    for order in all_orders:
        for each_menu in order:
            for each_menu_price in menus_dict:
                if each_menu_price == each_menu:
                    order_prices.append(menus_dict[each_menu])
        all_prices.append(order_prices)
        order_prices = []
    return all_prices

prices = assign_price(menus, menu_prices)
print(prices[0:28])
```

The next thing is to create a list of staff names and assign each name to each order based on the following criteria:

1. The first staff in the list works only on Mon, Wed, and Fri.
2. The second staff works on Tue and Thu.
3. The last staff works only on weekend.

```python
#Create a list of staff names
staff_names = ["Rambo", "Penguin", "Medse"]

#Turn the list of date strings into datetime object in the format of short-weekday
date_objs = [dt.datetime.strptime(date, "%d-%m-%Y").strftime("%a") for date in dates]
print("Below are 10 examples of the dates list")
print(date_objs[0:10])
print("\n")
#Check if the length of date_objs is equal to dates
[print("The length is", len(date_objs), "which is equal to", len(dates)) if len(date_objs) == len(dates) else "There is something wrong in the code!"]
```

```python
#Create a list of repeated short-weekday based on an order number each day
repeated_day_objs = sum([[d] * o for d, o in zip(date_objs, orders)] , [])
print(repeated_day_objs)
```

Create a new list of assigned staff name based on the criteria above

```python
#Create a dictionary in order to map staff names to specific days
assign_staff = {"Rambo": ["Mon", "Wed", "Fri"], "Penguin": ["Tue", "Thu"], "Medse": ["Sat", "Sun"]}

staff = []
for day in repeated_day_objs:
    for key, values in assign_staff.items():
        for value in values:
            if day == value:
                staff.append(key)
print(staff)
        
```

__Create a dataframe from the following lists__:

```python
#Create a dictionary using data from the lists above
d = {"dates": repeated_dates, "times": times, "orders": menus, "prices": prices, "staff": staff}
sales_data = pd.DataFrame(d)
sales_data.head()
```

## Clean The Data Using Python

Clean orders and prices columns by creating 2 new dataframes for splitting each order or price to the other column. After that assign the 2 new dataframes back to the original dataframe.

```python
#Create the second dataframe for splitting orders column
df2 = pd.DataFrame(sales_data["orders"].tolist()).rename(lambda x: "order_{}".format(x + 1), axis = 1)
df2.head()
```

```python
#Create the thord dataframe for splitting pricess column
df3 = pd.DataFrame(sales_data["prices"].tolist()).rename(lambda x: "price_{}".format(x + 1), axis = 1)
df3.head()
```

```python
#Assign the df2 and df3 back to sales_data
sales_data = pd.concat([sales_data, df2, df3], axis = 1)


sales_data_cleaned = sales_data[["dates", "times", "order_1", "order_2", "order_3", "order_4", "order_5", 
                                 "price_1", "price_2", "price_3", "price_4", "price_5", "staff"]]

sales_data_cleaned.head()
```

```python
#Replace all none values to NaN 
sales_data_cleaned = sales_data_cleaned.fillna(value = np.nan)
sales_data_cleaned.head()
```

Export the data to a csv file for later analysis 

```python
sales_data_cleaned.to_csv("happy_coffee_sales_cleaned.csv")
```
